```python
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = x_train.reshape(x_train.shape[0],-1)
x_train.shape
```

```python
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1) #for creating CNN layer
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
input_shape = (28, 28, 1)
```

# **Modelling**

```python
# convert class vectors to binary class matrices
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)
print(y_train)
```

```python
y_train.shape
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
batch_size = 128
num_classes = 10
epochs = 10

model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3),activation='relu',input_shape=input_shape, ))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
from keras.utils.vis_utils import plot_model
from IPython.display import Image

plot_model(model, to_file='model_plot.png', show_shapes=True, show_layer_names=True)
Image(filename='model_plot.png') 
```

```python
hist = model.fit(x_train, y_train,epochs=5,batch_size=10,verbose=1,validation_data=(x_test, y_test))
print("The model has successfully trained")
```

```python
print(hist.history.keys())
```

```python
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'cross validation'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'cross validation'], loc='upper left')
plt.show()
```

```python
model.save('mnist.h5')
print("Saving the model as mnist.h5")
```

```python
score = model.evaluate(x_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
y_pred = model.predict(x_test,verbose=1)

```

```python
y_pred = np.argmax(y_pred, axis=1) 
```

```python
correct_indices = list(np.where(y_pred == y_test)[0])
incorrect_indices = list(np.where(y_pred != y_test)[0])
```

```python
from sklearn.metrics import classification_report,confusion_matrix
report = classification_report(y_test,y_pred)
conf_matrix = confusion_matrix(y_test,y_pred)
print(report)
print(conf_matrix)
```
